```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('./Downloads/train.csv')
df_train = df_train.drop([1459])
df_train
```

```python
#Heat map to display null values in my data set
plt.figure(figsize=(14,10))
sns.heatmap(df_train.isnull(),cbar=False);
```

```python
#Data Types
df_train.info()
```

```python
#50% of data count
limit = df_train.shape[0]/2
print('If the count of null values goes beyond {}, remove those columns'.format(limit))
```

```python
#Displaying null values
df_train.isnull().sum().head(40)
```

```python
#Displaying null values
df_train.isnull().sum().tail(40)
```

```python
#Displaying all columns
df_train.head(10)
with pd.option_context('display.max_columns',100):
    display(df_train)
```

```python
#Verficiation of the int/cathegorical data
df_train['LotFrontage'].value_counts()
```

```python
#Defining Function for removal of null values
def cath_data(df,column):
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
#Defining Function for removal of null values
def int_data(df,column):
    df[column] = df[column].fillna(df[column].mean())
```

```python
#Defining Function for removal of null values
def drop_column(df,column):
    df.drop([column],axis=1,inplace=True)
```

```python
#Replacing Null values (Cathegorical Data)
cath_data(df_train,'MasVnrType')
cath_data(df_train,'BsmtCond')
cath_data(df_train,'BsmtQual')
cath_data(df_train,'BsmtExposure')
cath_data(df_train,'BsmtFinType1')
cath_data(df_train,'BsmtFinType2')
cath_data(df_train,'BsmtFinSF2')
cath_data(df_train,'BsmtUnfSF')
cath_data(df_train,'GarageType')
cath_data(df_train,'GarageFinish')
cath_data(df_train,'GarageQual')
cath_data(df_train,'GarageCond')
cath_data(df_train,'BsmtFullBath')
cath_data(df_train,'BsmtHalfBath')
cath_data(df_train,'KitchenQual')
cath_data(df_train,'Functional')
cath_data(df_train,'GarageCars')
cath_data(df_train,'MSZoning')
cath_data(df_train,'Utilities')
cath_data(df_train,'Exterior1st')
cath_data(df_train,'Exterior2nd')
cath_data(df_train,'SaleType')
cath_data(df_train,'Electrical')
```

```python
#Replacing Null values (Int Data)
int_data(df_train,'MasVnrArea')
int_data(df_train,'BsmtFinSF1')
int_data(df_train,'TotalBsmtSF')
int_data(df_train,'GarageYrBlt')
int_data(df_train,'GarageArea')
int_data(df_train,'LotFrontage')
```

```python
#Droping Null values
drop_column(df_train,'Alley')
drop_column(df_train,'FireplaceQu')
drop_column(df_train,'PoolQC')
drop_column(df_train,'Fence')
drop_column(df_train,'MiscFeature')
```

```python
#Verfying null values after data preprocessing
df_train.isnull().sum().head(40)
```

```python
#Verfying null values after data preprocessing
df_train.isnull().sum().tail(40)
```

```python
#last verification of null values in test data
plt.figure(figsize=(14,10))
sns.heatmap(df_train.isnull(),cbar=False,yticklabels=False);
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

cat_columns = ['Street','LandContour','Utilities','LotShape','LotConfig','LandSlope','Neighborhood','Condition1','Condition2',
              'BldgType','BldgType','RoofStyle','RoofMatl','RoofMatl','RoofMatl','ExterQual','MasVnrType','ExterQual',
              'ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
              'BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual','Functional','GarageType',
              'GarageFinish','GarageQual','GarageCond','PavedDrive','SaleType','SaleCondition','HouseStyle','Exterior1st',
              'Exterior2nd','MSZoning']

for i in range(len(cat_columns)):
    df_train[cat_columns[i]] = label_encoder.fit_transform(df_train[cat_columns[i]])  
```

```python
#List of all features consiting cathegorical values
columns=['Street','LandContour','Utilities','LotShape','LotConfig','LandSlope','Neighborhood','Condition1','Condition2',
              'BldgType','BldgType','RoofStyle','RoofMatl','RoofMatl','RoofMatl','ExterQual','MasVnrType','ExterQual',
              'ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
              'BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual','Functional','GarageType',
              'GarageFinish','GarageQual','GarageCond','PavedDrive','SaleType','SaleCondition','HouseStyle','Exterior1st',
              'Exterior2nd','MSZoning']
```

```python
#Verfying whether to use OneHotEncoding or get dummies function can also work
n = []
for col in range(len(columns)):
    print(columns[col],':', df_train[columns[col]].nunique(), 'labels')
    
print('As there are not many labels inside every column we can go ahead with get dummies for good accuracy')
```

```python
#To avoid comparison applying get dummies
df_train_new = pd.get_dummies(df_train,columns=['Street','LandContour','Utilities','LotShape','LotConfig','LandSlope','Neighborhood','Condition1','Condition2',
              'BldgType','BldgType','RoofStyle','RoofMatl','RoofMatl','RoofMatl','ExterQual','MasVnrType','ExterQual',
              'ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
              'BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual','Functional','GarageType',
              'GarageFinish','GarageQual','GarageCond','PavedDrive','SaleType','SaleCondition','HouseStyle','Exterior1st',
              'Exterior2nd','MSZoning'])
```

```python
#The data is encoded for the machine learning algorithm
df_train_new.head(10)
with pd.option_context('display.max_columns',304):
    display(df_train_new)
```

```python
x_train = df_train.drop(['SalePrice'],axis=1)
y_train = df_train['SalePrice']
```

```python
#Using Random Forest Regressor for predicting 
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10)
regressor.fit(x_train,y_train)
```
